#Data Cleaning and Preprocessing


```python
import pandas as pd
import numpy as nm
```

```python
# Load the dataset
df = pd.read_csv('/content/01.Data Cleaning and Preprocessing.csv')
```

```python
# Identify and handle missing data with mean
df.fillna(df.mean(), inplace=True)
```

```python
df.isna().sum()
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
df["UCZAA"].value_counts()
```

```python
# Detect and address outliers
# To Remove outliers in a specific column
Q1 = df['ChipRate'].quantile(0.25)
Q3 = df['ChipRate'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['ChipRate'] < (Q1 - 1.5 * IQR)) | (df['ChipRate'] > (Q3 + 1.5 * IQR)))]
```

```python
df.describe()
```

```python
# Perform basic data transformations
# Min-Max scaling for a numerical feature
df['Y-Kappa'] = (df['Y-Kappa'] - df['Y-Kappa'].min()) / (df['Y-Kappa'].max() - df['Y-Kappa'].min())
df
```

```python
# Encode categorical variables (if needed)
# Consider One-hot encoding for a categorical column
df = pd.get_dummies(df, columns=['ChipRate'])
```

```python
df
```

```python
df1=df[(df['UCZAA']>1) | (df['Y-Kappa']>20)]
df1
```

```python
df2=df[(df['UCZAA']<=1) | (df['Y-Kappa']<=20)]
df2
```

```python
# Save the clean dataset
df.to_csv('clean_dataset.csv', index=False)
```
