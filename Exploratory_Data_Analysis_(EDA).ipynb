#Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset into a DataFrame
df = pd.read_csv('/content/INDIAvi (2).csv')
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns
```

```python
# Display summary statistics
df.describe()
```

```python
# Handle missing values
df.dropna(inplace=True)

#check for null values
pd.isnull(df).sum()

# Handle outliers
df.fillna(df.mean(), inplace=True)

# Data preprocessing (e.g., feature scaling, encoding categorical variables)
df = pd.get_dummies(df, columns=['views'])
```

```python
# Visualize data distributions using histograms

plt.hist(df['likes'], bins=10)
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Histogram of column_name')
plt.show()
```

```python
# Create a scatter plot to explore relationships between variables

plt.scatter(df['likes'], df['views'])
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Scatter Plot')
plt.show()
```

```python
# Create a box plot to visualize distributions and outliers
sns.boxplot(x='category_id', y='likes', data=df)
plt.xlabel('Category')
plt.ylabel('Numeric Value')
plt.title('Box Plot')
plt.show()
```

```python
# Compute correlations between numerical variables
correlation_matrix = df.corr()

# Visualize correlations using a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
ss= sns.countplot(x='views',data=df)
for bar in ss.containers:
  ss.bar_label(bar)
```

```python
ss= sns.countplot(data=df,x='category_id',hue='likes')
for bar in ss.containers:
  ss.bar_label(bar)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ss= sns.countplot(data=df,x='dislikes')
for bar in ss.containers:
  ss.bar_label(bar)
```

```python
fig1, ax1=plt.subplots(figsize=(12,7))
df.groupby('dislikes')['likes'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
